```python
import dill
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
import pandas as pd
```

```python
def format_trajectory(trajectory):
    states = trajectory['states']
    doms = [x['dom_elements'] for x in states]
    images = [x['screenshot'] for x in states]
    actions = [x['action'] for x in states]
    goals = [trajectory['utterance']]* (len(states)-1)
    df = pd.DataFrame()
    df['goals'] = goals
    df['doms'] = doms[:-1]
    df['images'] = images[:-1]
    df['actions'] = actions[1:]
    return df
df = format_trajectory(trajectory)
```

```python
plt.imshow(df.images.values)
```
